# Loading the Dataset

```python
import pandas as pd

# Load the Titanic dataset
titanic_df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

# Display the first few rows of the DataFrame
print(titanic_df.head())
```

# Exploring the Data

```python
# Display the column names
print(titanic_df.columns)
```

```python
# Check the data types of each column
print(titanic_df.dtypes)
```

```python
# Get summary statistics for numeric columns
print(titanic_df.describe())
```

```python
# Count the number of rows and non-null values in each column
print(titanic_df.info())
```

# Filtering Data

```python
# Filter passengers who survived
survived_passengers = titanic_df[titanic_df['Survived'] == 1]
survived_passengers
```

```python

# Filter passengers who were in first class
first_class_passengers = titanic_df[titanic_df['Pclass'] == 1]
first_class_passengers
```

```python
# Combine multiple conditions using logical operators like "&" (AND) and "|" (OR)
female_first_class = titanic_df[(titanic_df['Sex'] == 'female') & (titanic_df['Pclass'] == 1)]
female_first_class
```

# Grouping and Aggregation

```python
# Group by passenger class and calculate the average age for each class
average_age_by_class = titanic_df.groupby('Pclass')['Age'].mean()
print(average_age_by_class)
```

```python
# Group by passenger class and gender, and calculate the survival rate for each group
survival_rate_by_class_gender = titanic_df.groupby(['Pclass', 'Sex'])['Survived'].mean()
print(survival_rate_by_class_gender)
```

# Data Cleaning

```python
# Count the number of missing values in each column
print(titanic_df.isnull().sum())
```

```python
# Fill missing age values with the mean age
mean_age = titanic_df['Age'].mean()
titanic_df['Age'].fillna(mean_age, inplace=True)
```

```python
print(titanic_df.isnull().sum())
```

# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot for passenger class distribution
sns.countplot(data=titanic_df, x='Pclass')
plt.title('Passenger Class Distribution')
plt.show()
```

```python
# Scatter plot to visualize the relationship between age and fare
sns.scatterplot(data=titanic_df, x='Age', y='Fare', hue='Survived')
plt.title('Age vs. Fare (Survived vs. Not Survived)')
plt.show()
```

# Data Transformation

```python
# Create a new column 'FamilySize' by summing 'SibSp' and 'Parch'
titanic_df['FamilySize'] = titanic_df['SibSp'] + titanic_df['Parch']
titanic_df
```

```python
# Convert 'Sex' column to numeric (e.g., 0 for male, 1 for female)
titanic_df['Sex'] = titanic_df['Sex'].map({'male': 0, 'female': 1})
titanic_df
```

```python
# Convert 'Embarked' column to one-hot encoded representation
titanic_df = pd.get_dummies(titanic_df, columns=['Embarked'])
titanic_df
```

# Data Reshaping

```python
# Pivot the data to get the average age for each class and sex combination
age_pivot = titanic_df.pivot_table(index='Pclass', columns='Sex', values='Age', aggfunc='mean')
print(age_pivot)
```

```python
# Melt the data to transform it from wide to long format
melted_df = pd.melt(titanic_df, id_vars=['PassengerId', 'Name'], value_vars=['Age', 'Fare'])
print(melted_df.head())
```

# Data Sampling

```python
# Randomly sample 10 rows from the DataFrame
random_sample = titanic_df.sample(n=10, random_state=42)
print(random_sample)
```

# Handling Outliers

```python
# Calculate the z-score for 'Fare' column to detect outliers
from scipy import stats
z_scores = stats.zscore(titanic_df['Fare'])
outliers_indices = titanic_df.index[abs(z_scores) > 3]

# Remove outliers from the DataFrame
titanic_df = titanic_df.drop(outliers_indices)
```

# Merging and Joining DataFrames

```python
# Create a new DataFrame with information about passengers' cabins
cabin_df = pd.DataFrame({'PassengerId': [1, 2, 3, 4, 5],
                         'Cabin': ['C85', 'C123', 'E46', 'C23 C25 C27', 'B78']})

# Merge the cabin information with the main Titanic DataFrame
titanic_merged = pd.merge(titanic_df, cabin_df, on='PassengerId', how='left')
print(titanic_merged.head())
```

# Data Aggregation using agg()

```python
# Calculate multiple aggregate functions for 'Age' and 'Fare'
agg_functions = {'Age': 'mean', 'Fare': ['min', 'max', 'median']}
aggregated_data = titanic_df.agg(agg_functions)
print(aggregated_data)
```

```python
# Dropping Columns and Rows
# Drop the 'Cabin' column from the DataFrame
titanic_df.drop(columns=['Cabin'], inplace=True)

# Drop rows where 'Age' is missing
titanic_df.dropna(subset=['Age'], inplace=True)
```

```python
# Data Merging with Different Keys
# Create another DataFrame with information about passengers' names
name_df = pd.DataFrame({'ID': [1, 2, 3, 4, 5],
                        'Passenger_Name': ['John Doe', 'Jane Smith', 'Michael Johnson', 'Emily Brown', 'David Lee']})

# Merge the name information using different key columns
titanic_merged = pd.merge(titanic_df, name_df, left_on='PassengerId', right_on='ID', how='left')
print(titanic_merged.head())
```

```python
# Handling String Data
# Extract titles from 'Name' column using regular expressions
titanic_df['Title'] = titanic_df['Name'].str.extract(' ([A-Za-z]+)\.', expand=False)
print(titanic_df['Title'].value_counts())

# Replace values in 'Title' column using the 'replace()' method
titanic_df['Title'] = titanic_df['Title'].replace({'Mlle': 'Miss', 'Mme': 'Mrs', 'Ms': 'Miss'})
```

```python
# Basic Visualization with Pandas
# Plot a histogram for the 'Age' column
titanic_df['Age'].plot(kind='hist', bins=20, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()
```

```python
# Plot a bar chart for the count of passengers in each class
titanic_df['Pclass'].value_counts().plot(kind='bar', color='purple', alpha=0.7)
plt.title('Passenger Class Distribution')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()
```

```python
import seaborn as sns

# Create a box plot to visualize the distribution of 'Age' by 'Pclass'
sns.boxplot(data=titanic_df, x='Pclass', y='Age', hue='Sex')
plt.title('Age Distribution by Pclass and Sex')
plt.show()
```

```python
#Customizing Visualizations
# Create a scatter plot with custom markers and colors
titanic_df.plot(kind='scatter', x='Age', y='Fare', c='Pclass', cmap='viridis', s=50, alpha=0.8)
plt.title('Age vs. Fare (Colored by Pclass)')
plt.show()
```

```python
# Create a bar plot with custom colors and labels
survival_counts = titanic_df['Survived'].value_counts()
survival_counts.plot(kind='bar', color=['salmon', 'lightgreen'])
plt.title('Survival Counts')
plt.xticks([0, 1], ['Not Survived', 'Survived'], rotation=0)
plt.show()
```

```python
# Combining Pandas and Matplotlib/Seaborn
# Filter the DataFrame to include only passengers with ages above 20
adult_passengers = titanic_df[titanic_df['Age'] > 20]

# Create a bar plot for the count of adult passengers in each class
adult_passengers['Pclass'].value_counts().plot(kind='bar', color='lightblue', alpha=0.7)
plt.title('Adult Passenger Class Distribution')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.show()
```
