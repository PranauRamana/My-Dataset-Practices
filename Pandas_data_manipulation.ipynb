# Pandas - Series:


1.   One-dimensional labeled array that can hold data of any type (integer, float, string, etc.)
2.   It is similar to a NumPy array but with additional features like labeled indexing



```python
import pandas as pd
# Creating a Series from a list
data = [10, 20, 30, 40, 50]
labels = ['A', 'B', 'C', 'D', 'E']
series = pd.Series(data, index=labels)
series
```

```python
# Creating a Series from a dictionary
data_dict = {'A': 10, 'B': 20, 'C': 30, 'D': 40, 'E': 50}
series_from_dict = pd.Series(data_dict)
series_from_dict
```

# Data Cleaning and Preprocessing with Pandas


```python
# Identifying and Handling Duplicates
import pandas as pd

# Create a sample DataFrame with duplicates
data = {
    'ID': [1, 2, 3, 1, 4, 5, 2],
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice', 'David', 'Emma', 'Bob'],
    'Age': [25, 30, 22, 25, 27, 24, 30]
}

df = pd.DataFrame(data)
df
```

```python
# Identify duplicates
duplicates = df.duplicated()
print("Duplicates:\n", duplicates)
```

```python
# Drop duplicates based on all columns
df_cleaned = df.drop_duplicates()
print("\nDataFrame without duplicates:\n", df_cleaned)
```

```python
# Drop duplicates based on a specific column (e.g., 'ID')
df_cleaned_by_id = df.drop_duplicates(subset='ID')
print("\nDataFrame without duplicates based on 'ID':\n", df_cleaned_by_id)
```

```python
# Dealing with Outliers
import pandas as pd

# Create a sample DataFrame with outliers
data = {
    'Age': [25, 30, 22, 100, 27, 24, 30]
}
df = pd.DataFrame(data)
df
```

```python
# Calculate the z-scores for each data point in 'Age'
z_scores = (df['Age'] - df['Age'].mean()) / df['Age'].std()
z_scores
```

```python
# Identify outliers using a threshold (e.g., z-score > 2)
outliers = df[abs(z_scores) > 2]
print("Outliers:\n", outliers)
```

```python
# Remove outliers by filtering
df_cleaned = df[abs(z_scores) <= 2]
print("\nDataFrame without outliers:\n", df_cleaned)
```

```python
# Removing Irrelevant Data and Columns
import pandas as pd

# Create a sample DataFrame with irrelevant data
data = {
    'ID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Age': [25, 30, 22, 27, 24],
    'Country': ['USA', 'UK', 'Canada', 'USA', 'Australia']
}

df = pd.DataFrame(data)
df
```

```python
# Drop a specific column (e.g., 'Country') that is irrelevant
df_cleaned = df.drop(columns='Country')
print("DataFrame without 'Country' column:\n", df_cleaned)
```

```python
# Drop rows based on a condition (e.g., Age < 25)
df_cleaned = df[df['Age'] >= 25]
print("\nDataFrame without rows where Age < 25:\n", df_cleaned)
```

```python
# Data Imputation Techniques
import pandas as pd

# Create a sample DataFrame with missing data
data = {
    'ID': [1, 2, 3, pd.NA, 5],
    'Age': [25, pd.NA, 22, 27, 24]
}

df = pd.DataFrame(data)
df
```

```python
# Fill missing values with mean of the column
mean_age = df['Age'].mean()
df_imputed = df.fillna({'Age': mean_age})
print("DataFrame with missing values filled with mean:\n", df_imputed)
```

```python
# Forward fill missing values
df_ffill = df.fillna(method='ffill')
print("\nDataFrame with forward-filled missing values:\n", df_ffill)
```

```python
# Backward fill missing values
df_bfill = df.fillna(method='bfill')
print("\nDataFrame with backward-filled missing values:\n", df_bfill)
```

```python
# Data Transformation: Applying Functions and Mapping
import pandas as pd

# Create a sample DataFrame
data = {
    'ID': [1, 2, 3, 4, 5],
    'Age': [25, 30, 22, 27, 24]
}

df = pd.DataFrame(data)
df
```

```python
# Apply a function to a column (e.g., add 1 to each Age)
df['Age'] = df['Age'>5].apply(lambda x: x + 10000)
print("DataFrame with Age incremented by 1:\n", df)
```

```python
# Map values in a column to new values (e.g., replace IDs with names)
id_to_name = {1: 'Alice', 2: 'Bob', 3: 'Charlie', 4: 'David', 5: 'Emma'}
df['Name'] = df['ID'].map(id_to_name)
print("\nDataFrame with 'Name' column mapped from 'ID':\n", df)
```
