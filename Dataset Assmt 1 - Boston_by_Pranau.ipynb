Filtering the data

```python
# Filter Tax in Bostonhousing
BostonTax = df[df['TAX'] == 437]
BostonTax
```

```python
# Filter the age above 65
Age = df[df['AGE'] >= 65]
Age
```

Grouping


```python
# Group by ZN and Indus
Group = df.groupby('ZN')['INDUS'].mean()
print(Group)
```

Data Visualization


```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Bar plot Boston Housing
sns.countplot(data=df, x='LSTAT')
plt.title('Boston Housing Distribution')
plt.show()
```

```python
# Scatter plot to visualize the relationship between NOX and RM
sns.scatterplot(data=df, x='RM', y='NOX')
plt.title('NOXRM (RM VS NOX)')
plt.show()
```

Data Transformation



```python
# Create a new column 'PRANAU' by summing 'CRIM' and 'B'
df['PRANAU'] = df['CRIM'] + df['B']
df
```

Convert Numeric to char

```python
# Convert 'CHAS' column to numeric (e.g., 0 for N, 1 for Y)
df['CHAS'] = df['CHAS'].map({'N': 0, 'Y': 1})
df
```

Data Reshaping

```python
# Pivot the data
rm_pivot = df.pivot_table(index='INDUS', columns='CHAS', values='RM', aggfunc='mean')
print(rm_pivot)
```
