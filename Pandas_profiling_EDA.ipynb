```python
#importing libraries
import pandas as pd
import pandas_profiling as pp
import seaborn as sns

# Load the Titanic dataset from Seaborn
iris_df = sns.load_dataset('iris')
```

```python
# Step 1: Display the first few rows of the dataset
iris_df.head()
```

```python
# Step 2: Generate the Profile Report
profile = pp.ProfileReport(iris_df)
```

```python
# Step 3: Display the Report
profile.to_widgets()
```

```python
# Step 4: Save the Report as an HTML file
profile.to_file('iris_report.html')
```

# Customize the Profile Report.

```python
# Changing the Title
# Generate the profile report with a custom title
profile_custom_title = pp.ProfileReport(iris_df, title='Custom Iris Dataset Report')
```

```python
# Disabling Certain Analyses
# Generate a minimal report (quicker for large datasets)
minimal_report = pp.ProfileReport(iris_df, minimal=True)

# Exclude correlation analysis for specific variables (e.g., 'sepal_length' and 'sepal_width')
excluded_report = pp.ProfileReport(iris_df, vars={'correlation': {'calculate': False}})
```

```python
# Generate a random sample of 50 rows
sampled_df = iris_df.sample(n=50, random_state=42)
sampled_report = pp.ProfileReport(sampled_df)
```

```python
# Controlling Correlation Analysis
# Generate the profile report with Spearman correlation and a threshold of 0.3
correlation_report = pp.ProfileReport(iris_df, correlations={'spearman': {'calculate': True, 'threshold': 0.3}})
```

```python
# Save the profile reports as HTML files
profile_custom_title.to_file('iris_custom_title_report.html')
minimal_report.to_file('iris_minimal_report.html')
excluded_report.to_file('iris_excluded_report.html')
correlation_report.to_file('iris_correlation_report.html')
sampled_report.to_file('iris_sampled_report.html')
```
